## Read in First Instance

```python
#resource dataframe
df = pd.read_csv('j120_1.csv', encoding='utf-8', skiprows = 3, nrows = 4)
    
#label columns & rows
df.columns = np.arange(1,121)
df.rename(index={0:'r1',1:'r2',2:'r3',3:'r4'}, inplace=True)

#data cleaning
df[1] = df[1].str.strip('rr = [| ')
df[120] = df[120].str.strip(' |];')

#dummy variables
r0 = pd.DataFrame({'1': [0,0,0,0]})
#r121 = pd.DataFrame({'122': [0,0,0,0]})
r0.rename(index={0:'r1',1:'r2',2:'r3',3:'r4'}, inplace=True)
#r121.rename(index={0:'r1',1:'r2',2:'r3',3:'r4'}, inplace=True)

#change columns and join
df.columns = np.arange(2,122)
df = r0.join(df)
#df = df.join(r121)

df
```

```python
#duration dataframe
df_dur = pd.read_csv('j120_1.csv', delimiter = ' ,', skiprows = 2, nrows = 1, engine='python')

#data cleaning
df_dur.columns = ['col1']
df_dur['col1'] = df_dur['col1'].str.strip('d = [')
df_dur['col1'] = df_dur['col1'].str.strip('];')

#label duration columns & rows
df_dur.rename(index={0:'duration'}, inplace=True)
df_dur = df_dur.col1.str.split(",",expand=True,)
df_dur.columns = np.arange(1,121)

#dummy variables
r0 = pd.DataFrame({'1': [0]})
#r121 = pd.DataFrame({'122': [0]})
r0.rename(index={0:'duration'}, inplace=True)
#r121.rename(index={0:'duration'}, inplace=True)

#change columns and join
df_dur.columns = np.arange(2,122)
df_dur = r0.join(df_dur)
#df_dur = df_dur.join(r121)

df_dur
```

## Chain Together Instances

```python
x = 122

for i in data :

    cols = np.arange(x,x+120)
    
#resources
#label columns & rows; data cleaning
    df1 = pd.read_csv(i, encoding='utf-8', skiprows = 3, nrows = 4)
    df1.columns = cols
    df1.rename(index={0:'r1',1:'r2',2:'r3',3:'r4'}, inplace=True)
    df1[x] = df1[x].str.strip('rr = [| ')
    df1[x+119] = df1[x+119].str.strip(' |];')
    
#durations
#label duration columns & rows; data cleaning
    df2 = pd.read_csv(i, delimiter = ' ,', skiprows = 2, nrows = 1, engine='python')
    df2 = df2['n_tasks = 120;'].str.split(",",expand=True,)
    
    df2.columns = cols
    
    df2[x] = df2[x].str.strip('d = [')
    df2[x+119] = df2[x+119].str.strip('];')
    df2.rename(index={0:'duration'}, inplace=True)


#join resources
    df.T
    df1.T
    df = df.join(df1)

#join durations
    df2.T
    df_dur.T
    df_dur = df_dur.join(df2)
    
#increment for next instance
    x = x + 120
```

```python
x = 122

for i in data :

    new_index = np.arange(x,x+120)

#successors
    mycolumns = ['suc1','suc2','suc3','4']
    df3 = pd.read_csv(i, names= mycolumns,skiprows = 8, engine='python')
    #print(df3)
    
    df3 = df3.drop("4", axis=1)
    df3['suc1'] = df3['suc1'].str.strip('  {}];')
    df3['suc2'] = df3['suc2'].str.strip('}')
    df3['suc3'] = df3['suc3'].str.strip('}')

    #replace NaN, blanks, and None
    df3 = df3.fillna(fill)
    df3 = df3.replace('',fill)
    
    df3 = df3.set_index([new_index])
        
    #join successors
    df_suc = df_suc.join(df3.T)
    
    #increment for next instance
    x = x + 120
    
df_suc
```
